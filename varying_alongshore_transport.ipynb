<h3>Effect of Directional Wave Spreading on Alongshore Transport</h3>
Where features on the inner shelf (ridges, rock outcrops)  cause waves to refract, the directional spread in the wave spectra is likely to be greater shoreward of these features. This directional scatter, in turn, may reduce alongshore transport, even if the significant wave height, peak period, and average wave direction is unchanged. Barbaro et al. (2014; J. Coastal Res. 30(5):961-966) have proposed a CERC-like equation for estimating the effect of directional spread in alonghore transport, and conclude that increased scatter does, in fact, decrease tranport. If true, this has implications for coasts where inner shelf ridges or geology scatter waves non-uniformly along the coast. For example, on a straight coast with uniform offshore waves, transport rates will decrease behind wave-scattering features.
<h3>Function to Implement Barbaro et. al (2014)</h3>

```python
def LST(Hs,Tp,m,theta0,nf=21,nd=11):
    """
    Calculate longshore transport rate for synthetic deepwater wave spectrum.
    Based on  Barbaro et al., 2014
    
    Input:
        Hs - Significant wave height (m)
        Tp - Peak period (s)
        m  - Directional spreading parameter (smaller is narrower dist)
        theta0 - Angle of attack (degrees). 0 is shore paralell from west, 90 is shore normal
        nf - number of frequency bins (optional; default is 21)
        nd - number of direction bins (optional; default is 11; odd number is better)
        
    Returns: Q, delThetad, DSPR
        Q - Alongshore transport rate (not sure of units; eqn. 30)
        delThetad - Non-dimensional parameter that contains effect of spreading (eqn. 31)
        DSPR - Spreading parameter used in SWAN
        
    This code is preliminary and has not been checked for robustness or errors.
    """
    # some additional parameters that could be made variables
    g = 9.81 # gravitional accleration (m s-2)
    gams = 2.650 # specific weight of sediment []
    gama = 1.030 # specific weight of water []
    p =.4        # porosity []
    K = .32      # CERC value [?]
    db = Hs/0.7  # breaking depth [m]

    # Make directional spread using cos^m(theta)
    # restrict the directional spectra to +/-90
    thetabins = linspace(-90.,90.,nd+1)
    theta = (pi/180.)*(thetabins[0:nd]+0.5*diff(thetabins))
    ddr = (theta[1]-theta[0])   #width of directional bins (radians)
    Dtheta=cos(theta)**m
    # scale so sums to one
    Dtheta=Dtheta/sum(Dtheta)
    # calculate DSPR according to SWAN manual
    DSPR2 = (180./pi)**2 * sum( 2*sin( theta )**2 * Dtheta * ddr ) 
    #print "DSPR=",sqrt(DSPR2)
    
    # Make frequency spectrum (some parameters could be made variables)
    # X1 and X2 are shape factors for frequency spectra
    # (Hassleman et al., 1973 cited in Barbero et al., 2014)
    X1 = 3.3    # same as GAMMA in SWAN
    X2 = 0.08   # same as sigma in eqn. 6.3.15 in Holthuijsen (2007)

    # Min, max, and peak frequencies (s-1) in synthetic spectrum
    fmin = 1./50.
    fmax = 1./.5
    fp = 1./Tp

    # make log-spaced frequency bins, and calc. df and log-centered f
    fbinedges = logspace(log10(fmin),log10(fmax),num=nf+1)
    # freqencies centered in linear frequency space
    # fbincenters = fbinedges[0:nf]+0.5*diff(fbinedges)
    # frequencies centered in log frequency space (probably more correct)
    fbincenters = 10.**( (log10(fbinedges[0:nf])+log10(fbinedges[1:nf+1]))/2.)
    f = fbincenters
    ddf = diff(fbinedges)
    w = (f/fp) 
    
    # JONSWAP spectrum per Barbaro et al. (2014)
    # equivalent to eqn 6.3.15 in Holthuijsen (2007)
    gterm = exp( log(X1)*exp( -((w-1.)**2.)/(2.*X2**2.) ) )
    Ee = w**(-5.)*exp(-(5./4.)*w**(-4.))*gterm
    A = (Hs**2.)/ (g**2*( (pi/180.)*fp )**(-5.) *16.*sum(Ee*ddf) )
    Ef = A* g*g*( (pi/180.)*fp )**(-5.)*Ee
    
    # Matrix of spectral energy without directional variability
    E = tile(Ef.reshape(nf,1),(1,nd))  # Energy spectrum
    EE = tile(Ee.reshape(nf,1),(1,nd)) # Dimensionaless energy spectrum
    # matrices of df,dtheta
    df= tile(ddf.reshape(nf,1),(1,nd))
    Dth=tile(Dtheta,(nf,1))
    # matrices of f, theta
    f_grid = tile(f.reshape(nf,1),(1,nd))
    theta_grid = tile((180.*theta/pi),(nf,1))
    m0=sum(sum(E*Dth*df))
    Hs_check = 4.*sqrt(m0)
    #print "Hs_check: ",Hs_check,", Hs error: ",100.*(Hs-Hs_check)/Hs," percent."
    
    # calculate directional effect (eqn. 31)
    theta0_gr = (pi/180.)*( theta_grid + 90. - theta0 )
    delThetad = (sum(sum(EE*Dth*cos(theta0_gr)*sin(theta0_gr)*df))) \
    /(sum(sum(EE*Dth*df)))
    #print theta0,delThetad
    
    # first term in eqn. 30
    const = K*sqrt(g*db)/(2.*(gams/gama-1.)*(1.-p))
    # transport rate
    Q = const*Hs*Hs*delThetad
    return Q,delThetad,sqrt(DSPR2)
```

<h3>Simple Example with Alonghore "Bump" in Spreading</h3>

```python
# Test LST function
from scipy import stats
from scipy.stats import norm

# offshore wave parameters
Hs = 2.       # (m)
Tp = 10.      # (s)
theta0 = 15.  # (deg) (zero = shore normal)
# alongshore 
x = linspace(0.,200.,21)
m = 9.*ones_like(x)
# increase the spread (decrease m) in middle
xx = linspace(-3.,3.,5)
y=10*stats.norm.pdf(xx)
m[8:13] = m[8:13]-y

Q = zeros_like(x)
delTheta = zeros_like(x)
DSPR = zeros_like(x)
for i in xrange(len(m)):
    Q[i],delTheta[i],DSPR[i] = LST(Hs,Tp,m[i],theta0)

fig=plt.figure(figsize=(8.,6.))
plt.subplot(3,1,1)
plot(x,DSPR)
plt.ylim(10.,20.)
plt.ylabel('DSPR (deg)')
plt.text(150,18,'Directional Spreading')
plt.subplot(3,1,2)
plot(x,Q)
plt.ylabel('Q')
plt.ylim(.60,.75)
plt.text(150,.62,'Alongshore Transport')
plt.subplot(3,1,3)
plot(x[1:],diff(Q))
plt.ylim(-.1,.1)
plt.ylabel('dQ/dx')
plt.text(150,.05,'Gradient in Transport')
xlabel('Alongshore Distance (km)')
```
