```python
# Plot one of the Dauphin Is. lidar DEMS
import numpy
import gdal
from gdalconst import *
fp = r"D:\crs\proj\2013_BIER\Dauphin\lidar\\"
fn = "16RCU388346.img"
dataset = gdal.Open(fp+fn, GA_ReadOnly )
print 'Driver: ', dataset.GetDriver().ShortName,'/', \
    dataset.GetDriver().LongName
print 'Size is ',dataset.RasterXSize,'x',dataset.RasterYSize, \
    'x',dataset.RasterCount
print 'Projection is ',dataset.GetProjection()
    
geotransform = dataset.GetGeoTransform()
if not geotransform is None:
    print 'Origin = (',geotransform[0], ',',geotransform[3],')'
    print 'Pixel Size = (',geotransform[1], ',',geotransform[5],')'
```

```python
topo=dataset.ReadAsArray()
#i, j = np.where(topo>0)
#topo = topo[min(i):max(i)+1, min(j):max(j)+1]
topo[topo<=-10.] = np.nan
print topo.shape
fig = plt.figure(frameon=False)
plt.imshow(topo, cmap=cm.BrBG_r)
#plt.pcolormesh(flipud(topo[500:1000,500:1000]), cmap=cm.BrBG_r)
plt.axis('off')
cbar = plt.colorbar(shrink=0.75)
cbar.set_label('meters')
plt.savefig(fp+'dauphin.png', dpi=300, bbox_inches='tight')
plt.show()
```
