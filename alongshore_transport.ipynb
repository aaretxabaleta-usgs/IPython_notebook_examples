```python
# Make a fake directional spectrum
Hs = 2. # Significant wave height (offshore incident waves)
Tp = 10.# Peak period (offshore incident waves)
nf = 11 # Number of frequency bins
nd = 5 # Number of directional bins (odd number is better)
m =  7. # Larger m = narrower distribution
```

<h3>Directional Spreading</h3>
SWAN uses the distribution function $cos^m(\theta-\theta_0)$. The (one-sided) directional spreading function is computed (eq. 6.3.21 in Holthuijsen, 2007) 

```python
# restrict the directional spectra to +/-90
thetabins = linspace(-90.,90.,nd+1)
theta = (pi/180.)*(thetabins[0:nd]+0.5*diff(thetabins))
ddr = (theta[1]-theta[0])   #width of directional bins (radians)
# make cos^m directional spread, where m is same as used in SWAN
# could put alternative spreading function in here
Dtheta=cos(theta)**m
# scale so sums to one
Dtheta=Dtheta/sum(Dtheta)
# calculate DSPR according to SWAN manual
DSPR2 = (180./pi)**2 * sum( 2*sin( theta )**2 * Dtheta * ddr ) 
print "DSPR=",sqrt(DSPR2)
plt.plot(180.*theta/pi,Dtheta)
plt.title('Directional Spreading')
plt.xlabel(r'Direction from $\theta _0$ (deg)')
plt.ylabel(r'$D(\theta)$ (1/rad)')
ts = "m = {0}, DSPR = {1:5.2f}".format(m,sqrt(DSPR2))
plt.text(0.3*max(180.*theta/pi),.8*max(Dtheta),ts)
```

```python
# frequency spectrum

# X1 and X2 are shape factors for frequency spectra (Hassleman et al., 1973 cited in Barbero et al., 2014)
X1 = 3.3    # might be same as GAMMA in SWAN
X2 = 0.08

# Min, max, and peak frequencies (s-1) in synthetic spectrum
fmin = 1./50.
fmax = 1./.5
fp = 1./Tp

# make log-spaced frequency bins, and calc. df and log-centered f
fbinedges = logspace(log10(fmin),log10(fmax),num=nf+1)
# freqencies centered in linear frequency space
#fbincenters = fbinedges[0:nf]+0.5*diff(fbinedges)
# frequencies centered in log frequency space (probably more correct)
fbincenters = 10.**( (log10(fbinedges[0:nf])+log10(fbinedges[1:nf+1]))/2.)
f = fbincenters
ddf = diff(fbinedges)
w = (f/fp) 
# Dimensionless JONSWAP spectrum per Barbaro et al. (2014)
gterm = exp( log(X1)*exp( -((w-1.)**2.)/(2.*X2**2.) ) )
Ee = w**(-5.)*exp(-(5./4.)*w**(-4.))*gterm              # Eqn 28
# Phillips parameter from eqn 29
A = (Hs**2.)/ (g**2*( (pi/180.)*fp )**(-5.) *16.*sum(Ee*ddf) )
# Dimensional spectrum 
Ef = A* g*g*( (pi/180.)*fp )**(-5.)*Ee                  # Eqn 27

# same spectrum using  6.3.16 in Holthuijsen (2007) as a check
gamma = X1 # Yes
siga = X2  # Yes
alphaj = A # No...see below
gtermj = gamma**exp( -0.5* ((f/fp-1.)/siga)**2. )
# These match, which confirms the gamma == X1, siga == X2
#print "gterm, gtermj: ",gterm,gtermj
alphaj = 1
Ej = alphaj*g*g*(2.*pi)**(-4.)*f**(-5)*exp( (-5./4.)*(f/fp)**(-4.) )*gtermj
alphaj = (Hs**2.)/( 16.*sum(Ej*ddf) )
Ej = alphaj*Ej

fig=plt.figure(figsize=(8.,6.))
plt.subplot(2,2,1)
plt.plot(f/fp,Ef/max(Ef))
plt.plot(f/fp,Ej/max(Ej),'--r')
plt.ylim([1.e-6, 1])
plt.xlim([0., 2.5])
plt.title('Norm. JONSWAP Spectrum')
plt.ylabel('E(f)/E(fp)')
plt.xlabel('f/fp')

plt.subplot(2,2,2)
plt.plot(f,Ef)
plt.plot(f,Ej,'--r')
#plt.ylim([1.e-6, 1])
#plt.xlim([0., 2.5])
plt.title('Dimensional JONSWAP Spectrum')
plt.ylabel('E(f)')
plt.xlabel('f')
```
