```python
# fd_transport_limiter
#
# poached from this really neat Matlab example:
"""
%MIT18086_FD_TRANSPORT_LIMITER
%    Solves the transport equation u_t+cu_x=0 by finite
%    difference methods. Of interest are discontinuous initial
%    conditions. Implemented methods are (1) upwind,
%    (2) Lax-Friedrichs, (3) Lax-Wendroff, (4) Lax-Wendroff
%    with van Leer flux limiter, and (5) Lax-Wendroff with
%    Superbee flux limiter.

% 03/2007 by Benjamin Seibold
%            http://www-math.mit.edu/~seibold/
"""
def diffl(x):
    return np.append([0.],diff(x))

def diffr(x):
    return np.append(diff(x),[0.])

def sumr(x):
    return np.append(x[0:-1]+x[1:],2.*np.array(x[-1]))

def ic(x):
    """initial condition function"""
    return np.logical_and(x>-.7,x<-.2)*.8+np.logical_and(x>-.9,x<-.6)*.2

def limiter(r,method=1):
    """flux limiter"""
    if method == 1 : # van Leer
        phi = (abs(r)+r)/(1.+abs(r))
    elif method == 2 : # Superbee
        a = np.ones_like(r)
        phi = np.maximum(0.*a,np.maximum(np.minimum(a,2.*r),np.minimum(r,2.*a)))
    else :
        print "no method",method
    return phi

n = 160                           # number of space gridpoints
dt = 4e-3                         # time step
tf = 1.                           # final time
c = 1.1                           # speed of transport
method = 3
x = np.linspace(-1.,1.,n,endpoint=True)
h = x[1]-x[0]
u = ic(x)
u0 = u

for tn in arange(1,ceil(tf/dt)):
    theta = diffl(u)/(diffr(u)+3e-14);
    if method == 1:
        name = 'upwind'
        f = c*u;
    elif method == 2:
        name = 'Lax-Friedrichs'
        f = c*sumr(u)/2-h/2/dt*diffr(u)
    elif method == 3:
        name = 'Lax-Wendroff'
        f = c*sumr(u)/2-c**2*dt/2/h*diffr(u)
    elif method ==4:
        name = 'Lax-Wendroff von Leer'
        phi = limiter(theta,1)
        f = c*u+c/2*(1-c*dt/h)*diffr(u)*phi;
    elif method ==5:
        name = 'Lax-Wendroff Superbee'
        phi = limiter(theta,2)
        f = c*u+c/2*(1-c*dt/h)*diffr(u)*phi
    else:
        print "bad method"
                            
    u = u-dt*diffl(f)/h
                            
    if (fmod(tn,30.) <= 1.e-4):
        plot(x,u0,'b:',x,ic(x-c*tn*dt),'b-',x,u,'r.-')
    
    plt.title(name)
```
